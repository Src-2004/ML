#In this line we calculate the intercept in mathe matical equation y=mx+b

```python
regression=LinearRegression(fit_intercept=True)
```

#Calculates the coefficients (m) and intercept (b) for the linear regression
#equation y=mX+b

```python
regression.fit(X_train,y_train)
```

#regression.coef means the cofficient of the features in the linear regresion y=m1x1+m2x2+...b
#it predict the y-value when all features (x1,x2,..)are 0

```python
print('Linear Model Coefficient (m):',regression.coef_)
print('Linear Model Intercept (b):',regression.intercept_)
```

#let's evoluting the model performence by calculating the residual sum of squares and the explained variance score(R^2).
#calculate the mean apsolute error , mean sequence and the Root sequence Error .

```python
from sklearn import metrics
print('MAE:',metrics.mean_absolute_error(y_train,predections))
print('MSE:',metrics.mean_squared_error(y_train,predections))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_train,predections)))
```
